```python
# TUPLURI

# Tuplurile sunt similare cu liste, dar sunt imutabile, adica nu pot fi modificate.

# Tuplurile sunt folosite pentru a stoca mai multe elemente intr-o singura variabila.

# Tuplurile sunt create prin adaugarea de paranteze rotunde la elemente:

# Exemplu

thistuple = ("apple", "banana", "cherry")

print(thistuple)

# Nota: Daca tuplul contine un singur element, trebuie sa adaugati o virgula dupa element, altfel Python nu va intelege ca este un tuplu.

# Exemplu

thistuple = ("apple",)

print(type(thistuple))

# NU este un tuplu

thistuple = ("apple")

print(type(thistuple))

# Functii de tupluri

# Python are doua functii integrate care pot fi utilizate pentru un tuplu:

# len() - Returneaza lungimea tuplului

# max() - Returneaza elementul maxim din tuplu

# min() - Returneaza elementul minim din tuplu

a = tuple("abc")
b = tuple([1,2,3])

tuple1 = ("a", "b" , "c")
item1, item2, item3 = tuple1

print(item1)
print(item2)
print(item3)

item1, *item2 = tuple1

print(item1)
print(item2)

print(tuple1 + tuple1)
```
