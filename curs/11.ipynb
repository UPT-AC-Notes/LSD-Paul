```python
'''
LOGICA PROPOZITIONALA

c, !c - literali - formule atomice


conjunctie - or - v

disjunctie - and - ^

forma normala conjunctiva (FNC) - conjunctie de disjunctii (in interiorul parantezelor sunt disjunctii de literali, iar in exterior este conjunctie de clauze)

forma normala disjunctiva (FND) - disjunctie de conjunctii (in interiorul parantezelor sunt conjunctii de literali, iar in exterior este disjunctie de clauze)



FNC

1. ducem negatiile inauntru, pana ajungem la literali - regula De Morgan

2. eliminam dublele negatii

3. inlocuim implicatiile ------ p -> q = !p v q

4. ducem disjunctia inauntru, pana ajungem la literali - regula distributivitatii

'''
```

```python
'''

Cum determinam daca o formula este realizabila?

- algoritm folosit in rezolvarea problemelor



SAT-problem / satisfiabilitate

- atribuire de valori de adevar pentru literali, astfel incat formula sa fie adevarata


- simplificam problema

- forma normala conjunctiva - daca avem un literal singur intr-o clauza, atunci clauza este adevarata daca literalul este adevarat


- daca ajungem la o clauza vida, atunci formula este nerealizabila

- daca nu mai putem simplifica, atunci formula este realizabila
    - alegem o variabila si ii atribuim valoarea de adevar 1, apoi simplificam formula

'''
```

```python
'''

DEMONSTRATIE VS CONSECINTA LOGICA


MODUS PONENS


Axiome - tautologie

A1: p -> (q -> p)
A2: (p -> (q -> r)) -> ((p -> q) -> (p -> r))
A3: (!q -> !p) -> (p -> q)

Demonstratia 

- insiruire de afirmatii, fiecare afirmatie fiind o tautologie sau o formula demonstrata prin aplicarea unei reguli de inferenta asupra unor formule anterioare


MODUS PONENS

a    a -> b
-------------
      b


a - adevarat
a -> b - adevarat

------> b - adevarat


alte moduri de deductie


- modus tollens


a -> b    !b
-----------------
      !a


- generalizare

    a
---------
   a v b
   
   
- specializare
- eliminare
- tranzitivitate




ipoteza - propozitie care nu este demonstrata, dar care este folosita in demonstratie (adevarata)


H - multime de ipoteze

H |- p - p este demonstrata folosind ipotezele din H





CONSECINTA LOGICA

- o formula este consecinta logica a unui set de formule daca exista o demonstratie a formulei folosind formulele din set


H |= p - p este consecinta logica a lui H


- dam valori de adevar pentru literali, iar apoi verificam daca formula este adevarata



daca H este o multime de formule, p este un literal, si H |- p, atunci H |= p


daca H este o multime de formule, p este o formula, si H |= p, atunci H |- p

'''
```
